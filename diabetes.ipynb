```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

print("Semua modul berhasil diimpor.")
```

load dataset

```python
diabetes_dataset= pd.read_csv('diabetes.csv')
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset['Outcome'].value_counts()
```

```python
#memisahkan data dan label 
X = diabetes_dataset.drop (columns='Outcome',axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print (X)
print (Y)
```

standarisasi data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
print(standarized_data)
```

```python
X = standarized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

memisahkan data training dan data testing

```python
X_train , X_test, Y_train, Y_test = train_test_split(X,Y , test_size = 0.2, stratify= Y , random_state=2)
```

```python
print (X.shape, X_train.shape, X_test.shape)
```

membuat data latih menggunakan algoritma svm 


```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit (X_train, Y_train)
```

membuat model evaluasi untuk mengukur akurasi 


```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score (X_train_prediction, Y_train)
```

```python
print('Akurasi training adalah =', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score (X_test_prediction, Y_test)
```

```python
print('Akurasi training adalah =', test_data_accuracy)
```

membuat model prediksi


```python
input_data = (6,	148,	72,	35,	0,	33.6,	0.627,	50
)
input_data_as_numpy_aray = np.array(input_data)

input_data_reshape = input_data_as_numpy_aray.reshape(1,-1)

std_data = scaler.transform(input_data_reshape)
print(std_data)
prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0]== 0) :
    print ('Pasien tidak terkena diabetes ')
else :
    print ('Pasien terkena diabetes')
```
